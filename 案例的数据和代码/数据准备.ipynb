```python
# 1. 定义并发症ICD编码规则
complication_icds = [
    '250.4', '250.5', '250.6', '250.7', '250.8',
    '357', '362', '366', '403', '404', '433', '440', '441',
    '583', '585', '588', '707', '709', 'V497'
]

# 2. 生成【是否有并发症】标签列 has_complication
def has_complication(row):
    diag_list = [str(row['diag_1']).strip(),
                 str(row['diag_2']).strip(),
                 str(row['diag_3']).strip()]
    for diag in diag_list:
        # 跳过填充值0（原缺失位置）
        if diag == '0':
            continue
        for icd in complication_icds:
            if icd in diag:
                return 1
    return 0

data['has_complication'] = data.apply(has_complication, axis=1)

# 3. 删除 diag_1、diag_2、diag_3 三列诊断字段
drop_diag_cols = ['diag_1', 'diag_2', 'diag_3']
data.drop(columns=drop_diag_cols, axis=1, inplace=True, errors='ignore')

# 可选：查看标签分布
print("并发症标签分布：")
print(data['has_complication'].value_counts())

# 可选：预览处理后数据
pd.set_option('display.max_columns', None)
print("\n删除诊断列后的数据预览：")
print(data.head())
```

```python
import pandas as pd

# --------------------------
# 1. 读取数据（保持原始文本不变，不要提前fillna！）
# --------------------------
# data = pd.read_csv('diabetic_data.csv')  # 你的读取代码

# --------------------------
# 2. 先处理文本列的编码（此时数据还是原始文本）
# --------------------------
# 2.1 性别：女=0，男=1，未知/空=2
gender_map = {
    'Male': 1,
    'Female': 0
}
data['gender'] = data['gender'].map(gender_map).fillna(2).astype(int)

# 2.2 种族：白种人0、非裔1、亚裔2、西班牙裔3；Other/空值=4
race_map = {
    'Caucasian': 0,
    'AfricanAmerican': 1,
    'Asian': 2,
    'Hispanic': 3,
    'Other': 4
}
data['race'] = data['race'].map(race_map).fillna(4).astype(int)

# 2.3 年龄区间编码
age_map = {
    '[0-10)': 1,
    '[10-20)': 2,
    '[20-30)': 3,
    '[30-40)': 4,
    '[40-50)': 5,
    '[50-60)': 6,
    '[60-70)': 7,
    '[70-80)': 8,
    '[80-90)': 9,
    '[90-100)': 10
}
data['age'] = data['age'].map(age_map).fillna(0).astype(int)

# 2.4 最高血清血糖 max_glu_serum
glu_map = {
    'None': 0,
    '>200': 1,
    '>300': 2,
    'Norm': 3
}
data['max_glu_serum'] = data['max_glu_serum'].map(glu_map).fillna(0).astype(int)

# 2.5 糖化血红蛋白 A1Cresult
a1c_map = {
    'None': 0,
    '>7': 1,
    '>8': 2,
    'Norm': 3
}
data['A1Cresult'] = data['A1Cresult'].map(a1c_map).fillna(0).astype(int)

# 2.6 再次入院情况 NO→0  >30→1  <30→2
readmit_map = {
    'NO': 0,
    '>30': 1,
    '<30': 2
}
data['readmitted'] = data['readmitted'].map(readmit_map).fillna(0).astype(int)

# --------------------------
# 3. 再处理全局缺失值（此时文本列已经变成数字，不会被影响）
# --------------------------
# 把 ? 转为 NaN，再填充为 0
data = data.replace('?', pd.NA)
data = data.fillna(0)

# --------------------------
# 4. 原生预览（和你之前的格式完全一致）
# --------------------------
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 10)

data.head()
```
